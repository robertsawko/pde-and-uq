# Introduction
Instead of running expensive Monte Carlo experiments one can derive PDE describing the change in the probability density function for the model problem. Note that the one-point PDF $p(x, t; a)$ is a function of three different arguments: position $x$, time $t$ and velocity value $a$. The functional dependance is dropped for notational convenience. In Heyrim article (add reference!) the PDE for PDF takes the following form:

\begin{equation}
\frac{\partial p(t)}{\partial t}
+ \int_{-\infty}^a \frac{\partial p}{\partial x} \textrm{d}a'
+ a \frac{\partial p}{\partial x}
=
-\sigma f(x, t; \textbf{b})
\frac{\partial p}{\partial a}
-\nu \frac{\partial}{\partial a} \left[ \left\langle \frac{\partial^2 u}{\partial x^2} \bigg| u\right\rangle p_u(t) \right]
\end{equation}

where $\textbf{b}$ are the coefficients of the forcing term. $\left\langle \frac{\partial^2 u}{\partial x^2} \bigg| u\right\rangle$ is a conditional average that could be obtained from the sampling of the original 1D problem. It is a function of $x$ and $t$ and it captures the expected value of the second order term under the assumption of known $u$ value at $x$, $t$ position. Also this term can be neglected (?) as we're in the vanishing viscosity limit.

In the $x$ direction boundary conditions should be periodic. In the a direction I believe we can do Dirichlet BC with p=0. Let $x \in [0, 2\pi]$ and $a \in [-4, 4]$.

### Implementation notes
 I think the idea is to implement the integral term by including an additional variable and equation and solve a system of equations i.e. set
 \begin{equation}
 I = \int_{-\infty}^a \frac{\partial p}{\partial x} \textrm{d}a'
 \end{equation}
 and then the original equation can be represented as a system:
 \begin{equation}
 \frac{\partial I}{\partial a} = \frac{\partial p}{\partial x}\quad (1)
 \\
\frac{\partial p}{\partial t}
+ I
+ a \frac{\partial p}{\partial x}
=
-\sigma f(x, t; \textbf{b})
\frac{\partial p}{\partial a}
-\nu \frac{\partial}{\partial a} \left[ \left\langle \frac{\partial^2 u}{\partial x^2} \bigg| u\right\rangle p_u(t) \right] \quad (2)
\end{equation}

The equation Proteus solves is
\begin{equation}
\mathbf{u} = \left[ u^0,u^1,\ldots,u^{nc-1} \right]
\end{equation}
\begin{equation*}
m^i_t + \nabla \cdot \left(\mathbf{f}^i - \sum_k^{nc-1} \mathbf{a}^{i,k} \nabla \phi^k(\mathbf{u}) \right) + r^i + H^i(\nabla u)= 0 \qquad \mbox{for } i=0,\ldots,nc-1
\end{equation*}

- $m$ is "mass"
- $\mathbf{f}$ is advection
- $\mathbf{a}$ is diffution
- $\phi$ is potential
- $H$ is Hamiltonian
- $r$ is reaction

Looking at (1) and (2) above what is $m$, $f$, $H$? etc?

Let $u^0=I$ and $u^1=p$ and $\nabla = \left[\frac{\partial}{\partial x}, \frac{\partial}{\partial a}\right]$ (Note: we are associating $a$ with the $y$ spatial direction). Then we seem to have two alternatives: one with advective operator and second with hamitlonian operator.



## With Hamiltonian

Chris suggested an implementation based on the Hamiltonian rather than the advection operator. It seems to be related to some implementation maturity(?) of Proteus with these two different operators. 

**NOTE** What I didn't realizae at first is that Hamitlonian for equation $i$ can be a function of the gradient of all variables from the solution vector. Indeed the canonical form quoted above suggests that one variable is taken!

Back to the plot though. With the Hamiltonian for the first and second equation the equation will take the following form:


\begin{align}
m^0 &= 0 & m^1 &= p \\
r^0 &= 0 & r^1 &= I \\
H^0(\nabla I, \nabla p) &=[0, 1] \cdot \nabla I + [-1, 0] \cdot \nabla p & H^1(\nabla p) &= \left[a, \sigma f(x,t)\right] \cdot \nabla p   \\
\end{align}
and diffusion term is zero.

```python
import proteus.default_p as p
import numpy as np
from proteus.TransportCoefficients import TC_base
Profiling.verbose=True
Profiling.logLevel = 5

p.nd = 2

# I am assuming L is the side length. Can we set the origin?
p.L = (2 * np.pi, 8)


class KineticEqn(TC_base):
    def __init__(self, nu=0.0, sigma=0.05, fofx=lambda x, t: 0):
        nc=2
        mass={}
        advection={}
        diffusion={}
        potential={}            
        reaction={}
        hamiltonian={}        
        mass[0] = {0:'constant'}
        mass[1] = {0:'linear'}
        reaction[0] = {0:'constant'}
        reaction[1] = {1:'linear'}
        hamiltonian[0] = {0:'linear', 1: 'linear'}
        hamiltonian[1] = {1:'linear'}
        # This is telling how many arguments you've got! so essentially the
        # line belows says "hamiltonian for equation 0 is a linear function of
        # gradient of 0th variable and linear function of a gradient of 1st
        # variable."
        TC_base.__init__(self,
                         nc=nc,
                         mass=mass,
                         advection=diffusion,
                         diffusion=diffusion,
                         potential=potential,    
                         reaction=reaction,
                         hamiltonian=hamiltonian,
                         variableNames=['I', 'p'])
        self.nu = nu
        self.fofx = fofx
        self.sigma = sigma

    def evaluate(self, t, c):
        I = c[('u', 0)]
        p = c[('u', 1)]
        
        x = c['x'][..., 0]
        a = c['x'][..., 1]
        
        gradI = c['grad(u)', 0]
        gradp = c['grad(u)', 1]
        
        c[('m', 0)][:] = 0
        c[('m', 1)][:] = p
        c[('m', 1, 1)][:] = 1
        
        c[('H', 0)][:] = gradI[..., 1] - gradp[..., 0]
        # derivatives with respect  to _gradient_ on hamiltonian terms
        c[('dH', 0, 0)][..., 0] = 0
        c[('dH', 0, 0)][..., 1] = 1.0
        c[('dH', 0, 1)][..., 0] = -1.0
        c[('dH', 0, 1)][..., 1] = 0
        
        c[('H', 1)][:] =\
            a * gradp[..., 0] + self.sigma * self.fofx(x, t) * gradp[..., 1]
        c[('dH', 1, 1)][..., 0] = a
        c[('dH', 1, 1)][..., 1] = self.sigma * self.fofx(x, t)

        c[('r', 0)][:] = 0

        c[('r', 1)][:] = I
        c[('dr', 1, 0)][:] = 1

```

```python
class pdf_IC:
    def __init__(self, b):
        self.b = b
        # Treat the constant jointly
        self.C = 5.0 / np.pi**2 * np.exp(-50.0 * b**2 / np.pi**2)

    def uOfXT(self, x, t):
        a = x[1] - 4
        # print a
        return self.C * np.exp(-(a - np.sin(x[0]))**2 / 2)


class I_IC:
    def __init__(self, b):
        pass
        self.pi2 = np.pi**2
        self.C = -5 / self.pi2
        self.b = b

    def uOfXT(self, x, t):
        a = x[1] - 4
        return self.C * np.cos(x[0])\
            * np.exp(
                - (np.pi**2 * np.sin(x[0])**2
                    - 2 * self.pi2 * np.sin(x[0]) * a
                    + self.pi2 * a**2 + 100 * self.b**2)
                / (2 * self.pi2))

p.initialConditions = {0: I_IC(xi),
                       1: pdf_IC(xi)}
```
